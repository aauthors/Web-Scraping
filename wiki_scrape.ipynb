# Step 1: Downloading The Page

The first thing when gathering data from a webpage is to download it using one of the python libraries. The one im using will be the `requests` library. This library will make what's called a **GET** request to the web server to download the **HTML** contents of the page. 

After running the code for the request we end up with a **Response** object with a property called `status_code` to indicate
if we downloaded the page correctly *200*. It also has a `content` property - the HTML content. 

```python
import requests 

page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
print(page.status_code)
print(page.content)
```

## Beautiful Soup

Here i'll use the Beautiful Soup library for data extraction. 

- purpose of beautiful soup is to pull data out of HTML and XML files.
- requires an input (document or URL). 
- have other modules such as regex, lxml. 
- process data in CSV or JSON format. 

```python
from bs4 import BeautifulSoup
```

```python
# create soup object

soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify)
```

```python
list(soup.children)
```

```python
# find all occurrences of p in HTML

print(soup.find_all('p'))
```

```python
print('\n\n')

# return only text

print(soup.find_all('p')[0].get_text())

type(soup.find_all('p')[0].get_text())
```
